###Delete Outliers

```python
#Create function to check outliers
def outliers(df, ft):
  Q1 = df[ft].quantile(0.25)
  Q3 = df[ft].quantile(0.75)
  IQR = Q3-Q1

  lower_bond = Q1 - 1.5 * IQR
  upper_bond = Q1 + 1.5 * IQR

  ls = df.index[(df[ft] < lower_bond) | (df[ft]>upper_bond)]
  return ls
```

```python
#create index list from GMV and Revenue that has outliers
index_list = []
for feature in ['GMV', 'Revenue']:
  index_list.extend(outliers(df_drop, feature))
```

```python
index_list
```

```python
#Create function to remove outliers
def remove(df,ls):
  ls = sorted(set(ls))
  df = df.drop(ls)
  return(df)
```

```python
#show data that hasn't outliers
df_outliers = remove(df_drop, index_list)
df_outliers
```

## Correlation between Total Buyer and Total Revenue

```python
#Creating group by order date
df_viz5 = df_new.groupby(['order_date']).agg({"oder_detail_id": pd.Series.nunique, "buyer_id": pd.Series.nunique, "GMV" : np.sum, "Revenue" : np.sum})

#Choose table that using in this analysis
df_viz5.columns = ["total_transaction","total_buyer","total_gmv","total_revenue"]
df_viz5
```

```python
#defining variable
x = df_viz5['total_buyer']
y = df_viz5['total_revenue']

#create scatter plot
fig, ax = plt.subplots(figsize = (9, 6))
plt.xlabel('total buyer')
plt.ylabel('total revenue')
ax.scatter(x,y, alpha=0.7,color="mediumvioletred")

#create line
b, a = np.polyfit(x, y, deg=1)
xseq = np.linspace(0, 900, num=20)
ax.plot(xseq, a + b * xseq, color="purple", lw=2.5);

# Add Text watermark
fig.text(0.87, 0.15, 'Database X-Company', fontsize = 12,
         color ='grey', ha ='right', va ='bottom')
```

Berdasarkan chart disamping, terlihat bahwa garis yang terbentuk adalah garis lurus menaik. Artinya, ada korelasi positif antara total pembeli dan total pendapatan. Jadi, jika total pembeli meningkat maka total pendapatan juga meningkat
